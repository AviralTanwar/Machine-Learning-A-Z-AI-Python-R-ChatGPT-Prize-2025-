Importing The Dataset

```python
# Using Pandas to import the dataset
import pandas as pd
```

```python
# Importing The Dataset
df = pd.read_csv(r'C:\Users\avira\Documents\Restart_Skills_v2025\GitHub\Machine_Learning_AI_Python_ChatGPT_dataset_venv\datasets\Part 2 - Regression\simple_linear_regression\Salary_Data.csv')

# Splitting the dataset into the Independent Variable (X) and the Dependent Variable (y)
X = df.iloc[:, :-1].values
Y = df.iloc[:, -1].values
```

Splitting the Dataset

```python
# Splitting the Dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
```

```python

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=0)
```

Training the Simple Linear Regression Model on the Training Set

```python
# Importing the Linear Regression model from sklearn
from sklearn.linear_model import LinearRegression
```

```python
# Initializing the Linear Regression model
# We are just storing the model and have not trained it yet!!!
# We might need to tune the model by passing parameters to the LinearRegression() function
regressor = LinearRegression()

# Training the model
regressor.fit(X_train, Y_train)
```

Predicting the Test Set Results

```python
# Predicting the Test Set Results
# We are using the trained model to predict the values for X_test
y_pred = regressor.predict(X_test)
```

```python
print("Predicted Values: ", y_pred)
```

Visualizing The Training Set Results

```python
# Visualizing The Training Set Results and Test Set Results
import matplotlib.pyplot as plt
```

```python
# Using a Scatter plot to visualize the Training Dataset
plt.scatter(X_train, Y_train, color='red')

#Using a line plot to visualize the Linear Regression line
plt.plot(X_train, regressor.predict(X_train), color='blue')

plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# Using a Scatter plot to visualize the Testing Dataset
plt.scatter(X_test, Y_test, color='red')

# Using a line plot to visualize the Linear Regression line
# Should be same as the one we plotted for the Training set since it's the same model
plt.plot(X_train, regressor.predict(X_train), color='blue')

plt.title('Salary vs Experience (Testing Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

Making a Single Prediction

```python
x = int(input("Enter years of experience to predict salary: "))

salary = regressor.predict([[x]])


print(f"Predicted salary for {x} years of experience is: {salary[0]}")
```

Getting the final linear regression equation with the values of the coefficients

```python
print(regressor.coef_)
print(regressor.intercept_)

# Salary=9345.94×YearsExperience+26816.19
# y = 9345.94x + 26816.19
```
